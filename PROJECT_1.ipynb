**DATA DESCRIPTION**

The dataset contains information about order date, region, category, product name, sales, profit,
and quantity. It represents retail transaction-level data used for business performance analysis.




```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel("sales_data.csv")

df.head()
```

```python
df.info()
df.isnull().sum()
```

Dataset contain sales and profit details.

no major missing values.

data types verified for analysis.



```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

Overall business is profitable

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

West region generate more revenue.

some region show comparatively lower sales .

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

Sales peaked during certain months (likely festive or quater-end)

seasonal treands observed in business.


```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)

top_products.plot(kind='bar')
plt.title("Top 5 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.show()
```

Few product contribute significantly to total revenue.

business can focus marketing on this high performance items.

```python
category_profit = df.groupby('Category')['Profit'].sum()

category_profit.plot(kind='bar')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

Office supplies ans technology are highly profitable.
